```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from xgcm import Grid
```

```python
# Eq 0
ds_seq0= xr.open_dataset("/vortex1/jupyter-sanjanasatish/south_winds/south_old/linear_d2_prog__3618_192.nc", decode_times=False)

# Eq W
ds_seqw= xr.open_dataset("/vortex1/jupyter-sanjanasatish/south_winds/south_new/linear_d2_prog__3918_117.nc", decode_times=False)

# No curl
ds_curl0 = xr.open_dataset("/vortex1/jupyter-sanjanasatish/no_curl/south/linear_d2_prog__4957_222.nc", decode_times=False)

#eq 3
ds_seq3 = xr.open_dataset("/vortex1/jupyter-sanjanasatish/0.03_winds/linear_d2_prog__4917_232.nc", decode_times=False)

ds_seq0_prog = xr.open_dataset("/vortex1/jupyter-sanjanasatish/south_winds/south_old/linear_d2_cont__3608_195.nc", decode_times=False)
ds_seqw_prog = xr.open_dataset("/vortex1/jupyter-sanjanasatish/0.02new/linear_d2_cont__4308_020.nc", decode_times=False)
ds_seq3_prog = xr.open_dataset("/vortex1/jupyter-sanjanasatish/0.03_winds/linear_d2_cont__4907_235.nc", decode_times=False)
ds_scurl0_prog = xr.open_dataset("/vortex1/jupyter-sanjanasatish/no_curl/south/linear_d2_cont__4947_225.nc", decode_times=False)
```

```python
ds_stats1 = xr.open_dataset('/vortex1/jupyter-sanjanasatish/with_precip/ocean_geometry.nc', decode_times=False)
ds_stats1 = ds_stats1.rename({'lath' : 'yh','lonh' : 'xh','latq' : 'yq','lonq' : 'xq'}) #renaming the coords to align with the original dataset
grid1 = Grid(ds_curl0, coords= {'X':{'center':'xh','outer':'xq'}, 'Y':{'center':'yh','outer':'yq'}, 'Z':{'center':'zl','outer':'zi'}},periodic=['X'])
grid2 = Grid(ds_scurl0_prog, coords= {'X':{'center':'xh','outer':'xq'}, 'Y':{'center':'yh','outer':'yq'}, 'Z':{'center':'zl'}},periodic=['X'])

dz = grid1.diff(ds_curl0.zi, axis = "Z")
```

```python
dz
```

```python
ds_stats1
```

# Eq 0

```python
Vhbt = (ds_seq0_prog.vh).sum('zl')
Psi_0 = grid1.cumsum(Vhbt.isel(Time=-1), 'X', boundary='fill')
(Psi_0*(10**-6)).plot.contourf(levels=40,cmap='RdBu_r', vmax=70)
```

```python
Uhbt = (ds_seq0_prog.uh).sum('zl')
Psi_u0 =  grid2.cumsum(Uhbt.isel(Time=-1), 'Y', boundary='fill')
((-Psi_u0*(10**-6)+170)).plot.contourf(levels=50,cmap='RdBu_r',vmax=75)
```

```python
V_bt = (ds_seq0.v * dz).sum('zl')
Psi_bt = grid1.cumsum(V_bt.isel(Time=-1) * ds_stats1.dxCv.compute(), 'X', boundary='fill')
(Psi_bt*(10**-6)).plot.contourf(levels=40,cmap='RdBu_r', vmax =100)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, constrained_layout=True, figsize=(12, 10))

contour0=(-Psi_u0*(10**-6)+167).plot.contourf(ax=ax[0,0],levels=35,cmap='bwr', vmax=70, vmin=-70)
contour1=(+Psi_u2*(10**-6)-Psi_u0*(10**-6)).plot.contourf(ax=ax[0,1],levels=35,cmap='RdBu_r', vmax=30, add_colorbar=False)
contour2=(+Psi_u3*(10**-6)-Psi_u0*(10**-6)).plot.contourf(ax=ax[1,0],levels=35,cmap='RdBu_r', vmax=30, add_colorbar=False)
contour3=(+Psi_uc*(10**-6)-Psi_u0*(10**-6)).plot.contourf(ax=ax[1,1],levels=35,cmap='RdBu_r', vmax=30, add_colorbar=False)

contours=[contour0,contour1,contour2,contour3]

ax[0, 0].axhline(y=43.5)
ax[0, 1].axhline(y=43.5)
ax[1, 0].axhline(y=43.5)
ax[1, 1].axhline(y=43.5)

ax[0, 0].set_title("seq0", fontsize =16)
ax[0, 1].set_title("seq2 - seq0", fontsize =16)
ax[1, 1].set_title("scurl0 - seq0", fontsize =16)
ax[1, 0].set_title("seq3 - seq0", fontsize =16)

ax[0,0].tick_params(axis='both', labelsize=13)
ax[0,1].tick_params(axis='both', labelsize=13)
ax[1,0].tick_params(axis='both', labelsize=13)
ax[1,1].tick_params(axis='both', labelsize=13)

ax[0,0].set_xlabel('')
ax[0,1].set_xlabel('')
ax[1,0].set_xlabel("Longitude (degrees East)", fontsize = 15)
ax[1,1].set_xlabel("Longitude (degrees East)", fontsize = 15)


ax[0,0].set_ylabel("Latitude (degrees North)", fontsize = 15)
ax[0,1].set_ylabel('')
ax[1,0].set_ylabel("Latitude (degrees North)", fontsize = 15)
ax[1,1].set_ylabel('')

contour0.colorbar.set_label('Volume Transport (Sv)',fontsize=13)

# Add a common colorbar
cbar = fig.colorbar(contours[1], ax=ax,orientation='horizontal')
cbar.ax.tick_params(labelsize=14)
cbar.set_label('Volume Transport (Sv)', fontsize=15)  # Set the colorbar title

plt.show()

```
