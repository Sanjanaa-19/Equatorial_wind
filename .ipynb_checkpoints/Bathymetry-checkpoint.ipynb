```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from xgcm import Grid
```

```python
#new winds 
ds_w = xr.open_dataset("/vortex1/jupyter-sanjanasatish/winds/linear_d2_prog__2499_157.nc", decode_times=False)

#old winds 
ds_ow = xr.open_dataset("/vortex1/jupyter-sanjanasatish/old_winds_new_forcing/linear_d2_prog__2491_109.nc", decode_times=False)

#for land
geo = xr.open_dataset("/vortex1/jupyter-sanjanasatish/with_precip/ocean_geometry.nc", decode_times=False)
geo = geo.rename({'lath' : 'yh','lonh' : 'xh','latq' : 'yq','lonq' : 'xq'})
```

```python
zonal_mean1_w = ds_w.mean(dim = ["xh"])
zonal_mean1_ow = ds_ow.mean(dim = ["xh"])

zonal_mean2_w = ds_w.mean(dim = ["xq"])
zonal_mean2_ow = ds_ow.mean(dim = ["xq"])
```

```python
grid1= Grid(ds_ow, coords={'X':{'center':'xh','outer':'xq'},'Y':{'center':'yh','outer':'yq'}, 'Z':{'center':'zl','outer':'zi'}},periodic=['X'])
temp_ow=grid1.interp(ds_ow.temp,'Y',boundary='fill')

grid2= Grid(ds_w, coords={'X':{'center':'xh','outer':'xq'},'Y':{'center':'yh','outer':'yq'}, 'Z':{'center':'zl','outer':'zi'}},periodic=['X'])
temp_w=grid2.interp(ds_w.temp,'Y',boundary='fill')
```

# Isopycnals at the Easternmost Boundary

```python
fig,ax=plt.subplots(nrows=1,ncols=1, constrained_layout=True,figsize=(15,10))

new_temp_contour3=ds_w.temp.isel(xh = 79, Time=0).plot.contour(levels = 15, colors='r')
ax.clabel(new_temp_contour3,inline_spacing=-19,fontsize=12,colors='r')

ax.invert_yaxis()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1, constrained_layout=True,figsize=(15,10))

new_temp_contour3=ds_w.temp.isel(xh = 75, Time=0).plot.contour(levels = 30, colors='r')
ax.clabel(new_temp_contour3,inline_spacing=-19,fontsize=12,colors='r')

ax.invert_yaxis()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1, constrained_layout=True,figsize=(15,10))

old_temp_contour3=temp_ow.isel(xh = 75, Time=-1).plot.contour(levels = 25, colors='b')
ax.clabel(old_temp_contour3,inline_spacing=-19,fontsize=12,colors='b')

new_temp_contour3=ds_w.temp.isel(xh = 75, Time=0).plot.contour(levels = 25, colors='r')
ax.clabel(new_temp_contour3,inline_spacing=-19,fontsize=12,colors='r')

ax.invert_yaxis()
```
