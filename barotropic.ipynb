```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
from xgcm import Grid
```

```python
ds_seq0=xr.open_dataset('Files/seq0.nc')
ds_seq2=xr.open_dataset('Files/seq2.nc')
ds_seq3=xr.open_dataset('Files/seq3.nc')
ds_scurl0=xr.open_dataset('Files/scurl0.nc')
ds_eq0=xr.open_dataset('Files/eq0.nc')
ds_eq2=xr.open_dataset('Files/eq2.nc')
ds_eq3=xr.open_dataset('Files/eq3.nc')
ds_curl0=xr.open_dataset('Files/curl0.nc')
```

```python
ds_stats1 = xr.open_dataset('Files/ocean_geometry.nc', decode_times=False)
ds_stats1 = ds_stats1.rename({'lath' : 'yh','lonh' : 'xh','latq' : 'yq','lonq' : 'xq'}) #renaming the coords to align with the original dataset
grid1 = Grid(ds_seq0, coords= {'X':{'center':'xh','outer':'xq'}, 'Y':{'center':'yh','outer':'yq'}, 'Z':{'center':'zl','outer':'zi'}},periodic=['X'])
# grid2 = Grid(ds_scurl0, coords= {'X':{'center':'xh','outer':'xq'}, 'Y':{'center':'yh','outer':'yq'}, 'Z':{'center':'zl'}},periodic=['X'])

dz = grid1.diff(ds_seq0.zi, axis = "Z")
```

# Eq 0

```python
Uhbt = (ds_eq0.uh).sum('zl')
Psi_u0 = grid1.cumsum(Uhbt, 'Y', boundary='fill')

constant = grid1.cumsum(Psi_u0, 'X', boundary ='fill').isel(xh=0, yq=70)
```

```python
((-Psi_u0  + constant)*(10**-6)).isel(xq=10).plot()
plt.axhline(y=0)
plt.axvline(x=41.5)
plt.axvline(x=-33.5)
# plt.axvline(x=-32.5)
plt.axvline(x=-13.5)
```

```python
plt.figure(figsize=(9,8))
Uhbt = (ds_seq0.uh).sum('zl')
Psi_u0 = grid1.cumsum(Uhbt, 'Y', boundary='fill')

constant = grid1.cumsum(Psi_u0, 'X', boundary ='fill').isel(xh=0, yq=70)

contour = ((-Psi_u0  + constant)*(10**-6)).plot.contourf(levels=50, cmap='RdBu_r', vmax=75, add_colorbar=False)

plt.title('Barotropic Streamfunction - CTRL', fontsize=17)
plt.xlabel('Longitude (degrees East)', fontsize=17)
plt.ylabel('Latitude (degrees North)', fontsize=17)
plt.axhline(y=41.5)

plt.tick_params(axis='both', labelsize=16)
cbar = plt.colorbar(contour)
cbar.ax.tick_params(labelsize=15) 
cbar.set_label('Volume Transport (Sv)', fontsize=15)  

plt.show()
```
